```python
#마스크 조건문 예제~~!!!! very imfortant~~!!!!!!!!!!.......어려워

count0fmask = int(input("마스크 개수 입력 >> "))

mask8 = count0fmask // 8                 # 연산자//는 몫이다. %는 나머지
mask5 = count0fmask % 8 // 5             # 현재의 mask5은 남은 마스크가 5~7개 일 때만 적용 가능.

# 8개들이와 5개들이로 포장한 후 남은 마스크가 1~4개 일 때 !! 밑에
if count0fmask % 8 % 5 >= 1 : 
    mask5 += 1

print(f"8개들이 포장지 개수 : {mask8}")
print(f"5개들이 포장지 개수 : {mask5}")
```

```python
#학점 계산기

score = int(input("점수 입력 >> "))

if score >= 90 :
    grade = "A"

# if 문제 있는 조건식이 false일 경우에만 elif로 내려오기 때문에 사전에 90점 미만이라는 대전제가 깔려져서 내려오게 됨
elif score >= 80 :             #90 미만을 안 쓴 이유_ 이프엘리트구문, 첫번째줄부터 트루 펄스로 내려와서
    grade = "B"
elif score >= 70 :
    grade = "C"
elif score >= 60 :
    grade = "D"
else :
    grade = "F"
    
print(f"{score}점은 {grade}학점 입니다.")
```

```python
s = "Hi my name is SW^^"
search = input("검색할 문자 입력 >> ")
cnt = s.count(search)                        #숫자 세는 문자 카운트, s 문자열에 들어있는 search 문자의 개수가 몇 개인지.

if search in s :
    print(f"{search}는 {cnt}번 들어있네요 ^_^")
else :
    print(f"{search}는 들어있지 않네요 ㅠ_ㅠ")
```
