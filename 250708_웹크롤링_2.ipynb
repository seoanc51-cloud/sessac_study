### iframe 태그
- 다른 서버의 데이터나 리소스를 현재 서버의 웹 페이지에 포함시켜 보여주는 HTML 태그
- 웹 페이지이에 보여줄 데이터가 많은 경우 한번에 접근하려면 그만큼 소모되는 자원이 많아 속도가 느려질 수 있는데, 이를 방지하기 위해 iframe을 사용하여 다른 서버에서 데이터를 불러와서 한번에 보여줌.
- 또한 기업 측면에서 중요한 데이터들을 따로 관리하기도 하는데 그럴 때 관리되는 서버에서 값을 빼와서 보여주는 용도로도 사용함.
- iframe태그를 활용하여 데이터를 출력시켜주고 있다면 기존처럼 접근은 불가하고 해당 데이터가 있는 원천 서버의 주소로 접근해야함.

```python
# iframe 태그가 있는 원천 서버에 새롭게 요청_ 새롭게 링크 가져오기.
res_iframe = req.get("https://finance.naver.com/marketindex/exchangeList.naver", headers=U_A)
res_iframe
```

```python
soup_iframe = bs(res_iframe.text, "lxml")             # 객체화까지 다시 시켜주기.
```

```python
currency_iframe = soup_iframe.select("td.tit > a")
currency_iframe[0].text
```

```python
currency_iframe[0].text.strip()               #공백 없애주는 strip()
```

```python
# 통화기준율까지 크롤링하여 두 항목을 DF로 만들어보세요~!
```

```python
sale_iframe = soup_iframe.select("td.sale")
sale_iframe
```

```python
len(currency_iframe), len(sale_iframe)
```

```python
currency_list, sale_list = [], []

for i in range(len(currency_iframe)) :
    currency_list.append(currency_iframe[i].text.strip())
    sale_list.append(sale_iframe[i].text)
    
```

```python
len(currency_list), len(sale_list)
```

```python
finance_dic = {'통화명' : currency_list, '매매기준율' : sale_list}

finance_df = pd.DataFrame(finance_dic, index=range(1,59))
finance_df.index.name = "No"
finance_df
```

### 이제 1~10page의 리뷰를 모두 수집해보자 ~~~~~ 어려워
- page가 달라지면 어떻게 접근해야 할까?

```python
# tqdm : 반복문의 진행상황을 %로 알려주는 모듈
from tqdm import tqdm
```

```python
review_all_list = []
page_list = []

# 1~10 page별 반복 코드
for i in tqdm(range(1,11,1)) :
    # 리뷰 페이지별로 url이 다르기 때문에 페이지별로 요청을 해줘야 함.
    url = f"https://review4.cre.ma/bodyluv.kr/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-3&install_method=hardcoded&page={i}&parent_url=https%3A%2F%2Fbodyluv.kr%2Fproduct%2F%25EB%25B0%2594%25EB%2594%2594%25EB%259F%25BD-%25EB%25B2%25A0%25EA%25B0%259C-best-6%25EC%25A2%2585%2F439%2Fcategory%2F65%2Fdisplay%2F1%2F&product_code=439&secure_device_token=V2cfca5208c34cdf3dce549fa4a0d6abed687f5aa1f433a66dc994d8081aa6dad4ebe8578141be79b4a618ae79520d9272&widget_env=100&widget_style=#"
    res = req.get(url, headers=U_A)
    soup = bs(res.text, "lxml")
    reviews = soup.select("div.review_list_v2__message.js-collapsed-review-content.js-translate-text")

    
    for j in range(len(reviews)) :
        body_re = reviews[j].text.strip()
        if "등록된 네이버 페이 구매평" in body_re :
            body_re = body_re.replace(temp_re[-38:], "")
        review_all_list.append(body_re)
        page_list.append(f"Page{i}")


review_all_list
```

```python
len(review_all_list)
```

```python
review_dic = {"리뷰":review_all_list, "페이지": page_list}

review_df = pd.DataFrame(review_dic, index=(range(1,51)))
review_df.index.name = "No"
review_df
```
