#### input 태그
- 웹 상에서 사용자의 입력을 받기 위한 인터페이스를 제공하는 특수 태그
- opening태그만 있고 closing태그는 없는 형태
- 다양한 속성들을 가지고 있음
  - type : 입력되는 형식 지정
  - value : 서버로 전송할 값
  - disabled : 입력 비활성 상태

```python
# 활성화 사이즈와 비활성화 사이즈를 구분하기 위해 input태그 까지 접근하기
size = driver.find_elements(By.CSS_SELECTOR, '#optSizeSection > ul.items > li > input')
```

#### CSS_SELECT로 태그의 속성에 접근하는 방법
- 태그명[속성="~~"]

```python
temp = driver.find_element(By.CSS_SELECTOR, 'label[for="size_260"]')
temp.text
```

```python
temp = driver.find_element(By.CSS_SELECTOR, 'label[for="size_265"]')
temp.text
```

```python
# 15개의 사이즈에 접근하여 활성화 사이즈와 비활성화 사이즈를 구분하여 리스트에 저장하는 코드    ***** 복습 어려움
enabled_size, disabled_size = [], []

for i in size :
    input_id = i.get_attribute("id")            # label태그의 for속성에 들어갈 size_~~ 는 id값에 동일하게 있으므로 아래에서 활용하기 위해 추출
    is_disabled = i.get_attribute("disabled")   # disable는 구매가능한 사이즈인지 아닌지를 판별하기 위해 아래에서 활용 예정

    # label 태그의 for 속성을 정확히 지정하여 접근
    label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
    # 사이즈에 대한 텍스트만 추출
    label_text = label.text.strip()

     # is_disabled 는 'true'라는 문자열이나 None을 반환하는데 문자열에 1개 이상의 문자가 포함되어 있으면 True이고, Nonn은 False로 동작함
    if is_disabled :
        disabled_size.append(label_text)
    else :
        enabled_size.append(label_text)

print("구매 가능 사이즈 : ", enabled_size)
print("구매 불가능 사이즈 : ", disabled_size)
```

### 상위 4개 품목들의 정보수집

```python
# 상위 4개 품목의 상품명, 평점, 구매가능한 사이즈, 발볼넓이, 리뷰 개수(숫자만)를 출력하게 하는 반복문 작성하기
```

```python
driver = wb.Chrome()
driver.maximize_window()
time.sleep(2)
driver.get("https://www.nbkorea.com/product/featuredList.action?cateGrpCode=250100&cIdx=1385")
```

```python
# <쌤이랑>
for i in tqdm(range(4)) :
    imgs = driver.find_elements(By.CSS_SELECTOR, 'img.img_goods')
    imgs[i].click()
    time.sleep(2)

    title = driver.find_element(By.CSS_SELECTOR, 'h2.title')
    mean_rating_review = driver.find_element(By.CSS_SELECTOR, 'div.crema_product_reviews_score__container')

    size = driver.find_elements(By.CSS_SELECTOR, '#optSizeSection > ul.items > li > input')
    enabled_size, disabled_size = [], []
    for j in size :
        input_id = j.get_attribute("id")         
        is_disabled = j.get_attribute("disabled")   
        
        label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
        label_text = label.text.strip()
        
        if is_disabled :
            disabled_size.append(label_text)
        else :
            enabled_size.append(label_text)

    foot_width = driver.find_element(By.CSS_SELECTOR,'#optWidthSection > ul > li > label')
    cnt_reviews = driver.find_element(By.CSS_SELECTOR,'a.crema-product-reviews-count.crema-applied')

    print(title.text + " - " + mean_rating_review.text + " - " + foot_width.text + " - " +  cnt_reviews.text)
    print("구매가능 사이즈 : ", enabled_size)

    driver.back()
    time.sleep(2)
```

### 한 셀에 모든 코드 넣어서 동작 시키기
- 이번에는 상위 8개 품목에 대해 데이터프래임화 시키는 것 까지 만들어보자

```python
driver = wb.Chrome()
driver.maximize_window()
time.sleep(2)
driver.get("https://www.nbkorea.com/product/featuredList.action?cateGrpCode=250100&cIdx=1385")
time.sleep(2)

# 각 품목들의 세부정보를 담아줄 빈 리스트 생성
title_list, mean_rating_review_list, enabled_size_list, cnt_reviews_list, foot_width_list = [], [], [], [], []

try : 
    for i in tqdm(range(8)) :
        imgs = driver.find_elements(By.CSS_SELECTOR, 'img.img_goods')
        imgs[i].click()
        time.sleep(2)
        
        title = driver.find_element(By.CSS_SELECTOR, 'h2.title')
        title_list.append(title.text)
        mean_rating_review = driver.find_element(By.CSS_SELECTOR, 'div.crema_product_reviews_score__container')
        mean_rating_review_list.append(mean_rating_review.text)
        
        size = driver.find_elements(By.CSS_SELECTOR, '#optSizeSection > ul.items > li > input')
        enabled_size, disabled_size = [], []
        for j in size :
            input_id = j.get_attribute("id")         
            is_disabled = j.get_attribute("disabled")   
            
            label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
            label_text = label.text.strip()
        
        if is_disabled :
            disabled_size.append(label_text)
        else :
            enabled_size.append(label_text)
        enabled_size_list.append(enabled_size)   # 구매가능한 사이즈는 원래 리스트였으므로 text없이 그냥 리스트로 리스트에 추가
        
        foot_width = driver.find_element(By.CSS_SELECTOR,'#optWidthSection > ul > li > label')
        foot_width_list.append(foot_width.text)
        cnt_reviews = driver.find_element(By.CSS_SELECTOR,'a.crema-product-reviews-count.crema-applied')
        cnt_reviews_list.append(cnt_reviews.text[1:-1]+'개')
        
        driver.back()
        time.sleep(2)
except :
    print("데이터 수집 중 예기치 못한 문제 발생!")

print("수집된 데이터 개수 확인 :")
print(len(title_list), len(mean_rating_review_list), len(cnt_reviews_list), len(enabled_size_list), len(foot_width_list))

driver.quit()

# 데이터프레임 작업
newbal_dict = {"상품명":title_list, "리뷰 평점":mean_rating_review_list, "리뷰 개수": cnt_reviews_list, 
               "구매가능 사이즈":enabled_size_list, "발볼":foot_width_list}
newbal_df = pd.DataFrame(newbal_dict, index=range(1, len(title_list)+1))
newbal_df.index.name = "No"
pd.set_option('display.max_colwidth', None)
newbal_df
```
