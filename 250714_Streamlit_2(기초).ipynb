## 화면 레이아웃 설정
- **slider** : 왼쪽 사이드바 설정
- **columns** : 컬럼 공간 설정
- **tab** : 탭 공간 설정(특정 탭을 선택하면 화면에 해당 탭 내용만 보임)
- **expender** : 사용자가 클릭하여 확장하거나 축소할 수 있는 섹션 설정
- **empty** : 페이지의 특정 위치에 하나의 요소를 출력할 수 있는 독립적인 공간 설정
  - 새 요소를 추가하면 기존 요소가 대체되므로 가장 최신의 업데이트를 출력하는데 포커스를 맞춤
  - 단일 콘텐츠를 동적으로 업데이트하고 표시하는데 유용함
- **container** : 특정 위치에 여러개의 요소를 출력할 공간 설정
  - 여러 요소를 그룹화하고 나중에 업데이트하여 표시하는 방식
  - 새 요소를 추가하면 기존 요소를 바꾸지 않고 단순히 컨테이너에 추가하므로 누적된 정보 집합을 출력하는데 포커스를 맞춤.

### 사이드바 설정

```python
%%writefile module/myApp9.py

import streamlit as st

st.title('여기는 메인 화면 입니다.')

# 사이드바 설정
st.sidebar.title('여기는 사이드바 입니다.')
st.sidebar.header('타이틀 - 중')
st.sidebar.subheader('타이틀 - 소')
'---'    # 해당 선은 메인에 표시
st.sidebar.write('---')   # 해당 선은 사이드바에 표시

select = st.sidebar.selectbox('다음 중 한 가지 선택', ['부먹','찍먹'])
st.sidebar.write(f'Selectbox 선택 = {select}')
```

### with문을 활용한 사이드바, 컬럼 설정

```python
%%writefile module/myApp9.py

import streamlit as st

with st.sidebar :
    st.sidebar.title('여기는 사이드바 입니다.')
    st.sidebar.header('타이틀 - 중')
    st.sidebar.subheader('타이틀 - 소')
    '---'
    
    select = st.sidebar.selectbox('다음 중 한 가지 선택', ['부먹','찍먹'])
    st.sidebar.write(f'Selectbox 선택 = {select}')

# 컬럼 설정
#col1, col2, col3 = st.columns([1,2,2])    #만들고자 하는 컬럼 객체 수에 맞게 정수 입력
# 리스트로 컬럼의 너비 비율을 지정할 수 있음
col1, col2, col3, _  = st.columns([2,2,2,4]) 

with col1 :
    # divider : 구분선 색 지정
    st.header('컬럼_1', divider='red')
    st.image('data/img_4074.jpg')
    st.caption('뉴진스 사랑한다!')

with col2 :
    st.header('컬럼_2', divider='orange')
    st.image('data/다운로드.jpg')

with col3 :
    st.header('컬럼_3', divider='rainbow')
    st.image('data/다운로드으.jpg')
```
