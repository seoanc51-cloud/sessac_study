```python
elements = driver.find_elements(By.CSS_SELECTOR,'.card-container')
res = []
for element in elements[:]:
    sale_elements = element.find_elements(By.CSS_SELECTOR,'.sale')
    sales = []
    for sale_element in sale_elements:
        sales.append(sale_element.text.replace('\n',' '))
    sale_text = ''.join(sales)

    
    name_nodes = element.find_elements(By.CSS_SELECTOR, 'span.card_name')
    if len(name_nodes) > 0:
        card_name = " ".join(name_nodes[0].text.split())
    else:
        card_name = ""

    # 3) 카드사
    corp_nodes = element.find_elements(By.CSS_SELECTOR, 'span.card_corp')
    if len(corp_nodes) > 0:
        card_corp = " ".join(corp_nodes[0].text.split())
    else:
        card_corp = ""

    # 4) 결과 합치기
    res.append({
        "card_name": card_name,
        "card_corp": card_corp,
        "benefits_raw": sale_text
    })

# 확인
print(len(res))
for item in res[:]:
    print(item)
```

```python
user_prompt = f"""\
# 개발자 지시 
너는 신용카드 또는 체크카드를 추천해주는 도우미야.
사용자에게 맞는 카드를 친절하게 추천해주고 설명해주면 돼.
# 카드 데이터(context)
{context}
# 사용자 정보(input_data)
{input_data}
"""
```

```python
template = """\
# 개발자 지시
너는 신용카드 추천 도우미야.
카드 데이터(context)와 사용자 정보(input_data)를 참고해서 최적의 카드를 제안해줘.

출력은 반드시 JSON 형식으로 하고,
필드는 [추천카드, 추천이유] 두 가지여야 한다.

# 카드 데이터(context)
{context}

# 사용자 정보(input_data)
{input_data}
"""

# 프롬프트 동적 함수
def build_prompt(context, input_data):
    return template.format(context=context, input_data=input_data)

#  사용 예시
print(build_prompt("삼성카드: 커피 50% 할인", "커피를 매일 마셔요"))
print()

# 사용 예시 2
print(build_prompt("신한카드: 영화 30% 할인", "영화를 자주 봐요"))
print()

persona_1 = """
나이: 25세
성별: 여성
거주지: 서울 서초구 원룸(월세)
가족: 부모님은 지방 거주, 1인 가구
애완동물: 햄스터 1마리
직업: 스타트업 마케팅팀 사원(입사 1년 차), 블로거
생활패턴: 아침은 거르고, 점심은 회사 근처 식당, 저녁은 배달·간편식 또는 외식 위주
월 소득: 세후 약 230만 원
재산 상태: 예금 2000만 원, 적금 매달 30만 원
자동차: 없음, 대중교통 이용
건강: 영양제를 자주 사먹고 운동도 자주하지만 술을 많이 마셔서 주의 필요 
취미: 브이로그 촬영, 전시회·팝업 스토어 방문, 카페 탐방, 프리다이빙, 스쿠버다이빙
지출: 생활비 중 절반 이상이 취미활동, 나머지는 월세와 햄스터 용품
특이사항: 블로그로 외식 협찬을 자주 받음, 집에서 밥을 아예 안해먹음
"""

context = """
삼성카드: 커피 50% 할인
신한카드: 영화 30% 할인
현대카드: 대중교통 20% 할인, 편의점 10% 할인
우리카드: 배달앱 15% 할인
"""

print(build_prompt(context, persona_1))
```

```python
prompt = build_prompt(context, persona_1)

# 5) OpenAI API 호출
completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=200,
    temperature=0,
)

# 6) 결과 출력
print(completion.choices[0].message.content)
```
