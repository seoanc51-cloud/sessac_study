###반복문

#반복문_ 같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 문법,
#반복문 종류에는 while(반복 횟수가 명확하지 않을 때), for(반복문 횟수가 명확할 때)

```python
# while문 

num = 0
while num < 10:
    print("파이썬 최고!!")
    num += 1
```

```python
number = 1 
while number <= 3:
    print(number)
    number +=1
```

```python
#break  반복문 밖으로 나가는 명령

number = 1
while True:
    print(number)
    number += 1
    if number > 3:
        break
```

```python
# continue _ 반복문 시작으로 돌아가는 명령

num = 0
while num < 10:
    num += 1
    if num > 5:
        print("여기서 부터는 num이 증가하지 않습니다.")
        continue
    print(num)
```

```python
### 두개의 정수를 입력받는 문제

while True :   
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))

    if num1 == 0 and num2 ==0 :
        print("프로그램이 종료되었습니다.")
        break

    print(f"두 정수의 합 >> {num1 + num2} ")  
```

```python
#다이어트 관리프로그램  (내가 했당~~!!!)

number = 0

while True :
    num1 = int(input("현재 몸무게 : "))
    num2 = int(input("목표 몸무게 : "))
    number += 1
    
    if num1 == num2 :
        print(f"{num2} kg 달성!! 축하합니다!")
        break

    print(f"{number}주차 감량 몸무게 : {num1-num2}")

    
```

```python
#다이어트 관리프로그램  (쌤이 한거)

weight = int(input("현재 몸무게 :"))
goal = int(input("목표 몸무게 :"))
week = 1              

#목표 몸무게를 달성하지 못하는 경우에 반복적으로 다이어트를 진행해야 함

while weight > goal :
    weight_week = int(input(f"{week}주차 감량 몸무게 :"))
    weight -= weight_week  #매주 감량한 몸무게를 현재 몸무게에서 빼고 업데이트
    week += 1
print(f"{weight} kg 달성!! 축하합니다!")
```

```python
# up, down 게임 (내가한거..)

import random
num = random.randiant(1, 50)

while True:
    number = int(input("숫자를 입력하세요 >> "))

    if number == num :
    print("정답을 맞추셨습니다.")
    break

elif number > num:
print(f"{number}보다 작은 수 입니다.")

else number < num:
print(f"{number}보다 큰 수 입니다.")
```

```python
# up, down 게임 (쌤이랑..)

import random
num = random.randiant(1, 50)

while True :
    input_num = int(input("숫자를 입력하세요 >>"))

    if num > input_num :
        print(f"{input_num}보다 큰 수 입니다.")
    elif num < input_num :
        print(f"{input_num}보다 작은 수 입니다.")
    else :
        print("정답을 맞추셨습니다.")
        break
        
```

```python
#점수 등급별로 판별(내가한거,,)

python_score = [57, 86, 63, 92, 35, 79]
num = 1

for i in python_score :
    if i>=80 :
        print(f"{num}은 {python_score}점 이며, A등급 입니다.")
    elif i>= 60 :
        print(f"{num}은 {python_score}점 이며, B등급 입니다.")
    else :
        print(f"{num}은 {python_score}점 이며, C등급 입니다.")
```

```python
#점수 등급별로 판별 (쌤이랑 한거)

python_score = [57, 86, 63, 92, 35, 79]
num = 0

for i in python_score :
    if i>=80 :
        grade = "A"
    elif i>=60 :
        grade = "B"
    else :
        grade = "C"

    num += 1
    print(f"{num}번은 {i}점 이며, {grade}등급 입니다.")
    
```

```python
#enumerate 함수 활용

python_score = [57, 86, 63, 92, 35, 79]
num = 0

# enumerate : 순서가 있는 집합을 반복시 index와 값을 모두 변환시켜주는 함수 (index, 값)
for num, i in enumerate(python_score) :
    if i>=80 :
        grade = "A"
    elif i>= 60 :
        grade = "B"
    else :
        grade = "C"

    print(f"{num+1}번은 {i}점 이며, {grade}등급 입니다.")
```

```python
#첫번째 정수부터 두번째 정수까지 범위 안의 입력된 배수의 합을 출력하는 코드 *****

start = int(input("첫 번 째 정수 입력 >> "))
end =  int(input("두 번 째 정수 입력  >> "))
num = int(input("배수 입력 >> "))
total = 0

for i in range(start, end+1, 1) :
    if i % num == 0 :
        total += i
        
print(f"{num}의 배수의 합 : {total}")
```

#구구단 전체 출력하기. 2단부터 9단까지.- 반복문을 두번 사용해야한다. *****

```python
for dan in range(2, 10, 1) :
    for su in range(1, 10, 1) :
        print(f"{dan}*{su}={dan*su}", end =" \t")    # \t를 쓰먼 출력되는 값들 사이 간격을 맞춰준다.
    print()   #하나의 단이 끝난 후 개행을 위한 print문
```

```python
#짝수와 홀수의  개수를 출력. **(me)

list1 = [1, 30, 7, 9, 17, 3, 5, 23, 44]
total = 0

for i in list1 :
    if i % 2 ==0 :
        total += 1      
        print(f"짝수 {total}개")
    else :
        total += 1
        print(f"홀수 {total}개")
```

```python
#짝수와 홀수의 개수 출력 (쌤이랑)

list1 = [1, 30, 7, 9, 17, 3, 5, 23, 44]
even = 0     #짝수의 개수
odd = 0      #홀수의 개수

for i in list1 :
    if i % 2 ==0 :
        even += 1
    else :
        odd += 1

print(f"짝수{even}개, 홀수 {odd}개")
```

```python
#답을 연속으로 맞출수록 점수가 커지는 퀴즈 

OX = input("OX입력 >> ")

score = 0
streak = 0

for i in OX :
    if i == "O" :
        streak += 1
        score += streak
    else :
        streak = 0
        
print(f"{score}점")
```
