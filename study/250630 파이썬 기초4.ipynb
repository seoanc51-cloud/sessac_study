```python
# 더하거나 곱할 수 있는 함수 만들기 (매개 변수) ***
 
def calculator(choice, *args) :
    if choice == "add" :
        total = 0
        for i in args : 
            total += i
    elif choice == "mul" : 
        total = 1                   #값들끼리의 곱인 경우에는 초기값을 1오 설정해야 0이 나오지 않음.
        for i in args :
            total *= i
    return total
```

```python
calculator("add", 2,3,4)           #add 더하거나
```

```python
calculator("mul", 2,3,4)       #mul 곱할 수 있다.
```

```python
# 클래스 예제            ***

class FourCal :
    def setdata(self , num1, num2) :
        self.num1 = num1
        self.num2 = num2
    
    def add(self) :
        result = self.num1 + self.num2
        return result
    def sub(self) :
        result = self.num1 - self.num2
        return result

    def mul(self) :
        result = self.num1 * self.num2
        return result

    def div(self) :
        result = self.num1 / self.num2
        return result
```

```python
a = FourCal()
b = FourCal()

a.setdata(6,3)
b.setdata(3,8)

print(a.add())
print(b.add())
print(a.sub())
print(b.sub())
print(a.mul())
print(b.mul())
print(a.div())
print(b.div())
```

* 생성자: 객체가 생성될 때 자동으로 호출되는 메소드 _ 초기값을 설정해야 할 필요가 있는 경우 '__init__'사용.

```python
class FourCal :
    def __init__(self, num1, num2) :
        self.num1 = num1
        self.num2 = num2
    
    def add(self) :
        result = self.num1 + self.num2
        return result
    def sub(self) :
        result = self.num1 - self.num2
        return result

    def mul(self) :
        result = self.num1 * self.num2
        return result

    def div(self) :
        result = self.num1 / self.num2
        return result
```

```python
# 이름 성별 나이 입력 받아 쓰기 **

class Member :
    def __init__(self, name , gender, age):
        self.name = name
        self.gender = gender
        self.age = age
    def info(self) :
        print(f"이름 : {self.name}")
        print(f"성별 : {self.gender}")
        print(f"나이 : {self.age}")
```

```python
csw = Member()

#__int__ 으로 생성자를 지정해뒀기 때문에 항상 객체를 만들때 생성자에 드러가는 매개변수 만큼 설정해줘야 함.
```

```python
# Member 클래스를 통해 csw라는 객체를 만듦
csw = Member("성우", "남", "26")
```

```python
#만들어진 csw 객체가 info() 메소드로 행동함.
csw.info()
```

* 상속: 물려받는 것. 물려주는 클래스를 부모 클래스, 물려받는 클래스를 자식 클래스라 함.
자식 클래스는 부모 클래스의 메소드를 그대로 사용 가능, 재사용성을 극대화 한다.
  

```python
#상속 예제 ** 지수연산도 쓰시오.

class NewFourCal(FourCal) :
    def power(self) :
        result = self.num1**self.num2
        return result
```

```python
c = NewFourCal(3, 2)
```

```python
print(c.power())
print(c.add())
print(c.sub())
print(c.mul())
print(c.div())
```
