### BMI 지수 확인 및 인덱싱

#### 1. 데이터 로드

```python
# Loadtxt : numpy에서 txt파일을 로드해오는 함수
 # delimiter : txt파일에서 데이터를 구분하는 구분자를 설정해주는 인자

data = np.loadtxt('data/height_weight.txt', delimiter=',' )           #delimiter인 컴마,로 구분이 되어있는 데이터를 분류.
data

#키는 cm, 몸무게는 kg
```

#### 2. 데이터 속성 파악
- 배열 형태
- 배열 차원수
- 배열 전체 요소 개수

```python
print("배열 형태 :", data.shape)
print("배열 차원수 :", data.ndim)
print("배열 전체 요소 개수 :", data.size)
```

#### 3. 배열 데이터에서 키와 몸무게를 각각 변수에 담기

```python
#키
height = data[0]
#몸무게
weight = data[1]
```

```python
height
```

```python
weight
```

#### 4. BMI 지수 공식에 대입해보기

```python
BMI = (weight / (height * height))*10000
```

```python
BMI
```

#### 5. BMI지수가 23이상인 수치를 구해보세요~!

```python
BMI >= 23
```

```python
BMI[BMI >= 23]
```

#### 6. BMI지수가 23이상 25이하인 사람은 몇 명인지 구해보시오.
- numpy 배열 불리언 인덱싱에서는 파이썬 문법인 논리 연산자(and, or, not)를 사용할 수 없고, 비트 연산자(&, |, ~)를 사용하여 다중 조건을 판단함.
- numpy는 R언어의 연산자인 비트 연산을 기반으로 만들어졌기 때문

```python
BMI[(BMI>=23) & (BMI<=25)]

#비트연산자 
 # & : and연산
 # | : or연산
 # ~ : not연산
```

```python
len(BMI[(BMI>=23) & (BMI<=25)])
```

```python
BMI[(BMI>=23) & (BMI<=25)].size
```

### 영화 평점 데이터 분석 예제

1. 데이터 로드

```python
# genfromtxt : 텍스트 형식의 파일로부터 데이터를 로드해주는 함수.
movie = np.genfromtxt('data/ratings.dat', delimiter='::', dtype=np.int64)        #tab 키 누르면 자동완성.
movie

#사용자 아이디, 영화 아이디, 사용자가 매긴 영화평점, 입력시간
```

```python
print(movie.shape)
print(movie.ndim)
print(movie.size)
```

#### 2. 전체 영화 평점 평균 구하기

```python
ratings = movie[: , 2]   # [행의 범위, 열의 범위]
ratings
```

```python
ratings.size
```

```python
#전체 영화의 평점 평균
np.mean(ratings)
```

#### 3. 1번 사용자가 매긴 평점의 평균 구하기
- 사용자 아이디의 유니 값 확인
- 사용자 아이디 유니크 값 개수 확인
- 1번 사용자의 데이터에 접근(불리언 인덱싱)
-1번 사용자가 매긴 평점들의 평균 구하기

```python
# 평점을 매긴 사용자가 몇 명인지 알아보자!
user_id = np.unique(movie[:,0])     # uniqe : 유일한 값을 구해주는 함수
user_id
```

```python
user_id.size

# 사용자는 총 6040명
```

```python
#불리언 인덱싱(사용자의 아이디가 1번인 데이터들을 출력)
movie[movie[:, 0] == 1]
```

```python
# 1번 사용자의 모든 평점에 접근

movie[movie[:, 0] == 1][:, 2]
```

```python
movie[movie[:, 0] == 1][:, 2].mean()
```

```python
# 1번 사용자가 매긴 평점의 평균
np.mean(movie[movie[:, 0] == 1][:, 2])
```

- 전체 영화 평점이 3.58점이었는데 1번 사용자가 4.18점 인걸로 보아 1번 사용자는 평점을 후하게 주는 편!

#### 4. 각 사용자가 매긴 평점들의 평균 구하기

```python
# 비어있는 리스트를 만들고 각 사용자별 id와 평점의 평균값을 넣어보자.
user_mean_list = []

# 사용자가 1번부터 6040번까지이므로 6040번 반복
for i in user_id : 
   #사용자 별로 2번 인덱스 열의 값(평점)의 평균을 구함
    user_mean = movie[movie[:, 0] == i][:, 2].mean()
    user_mean_list.append([i, user_mean])
    
user_mean_list
```

```python
len(user_mean_list)
```

#### 각 영화가 받은 평점의 평균 구하기
