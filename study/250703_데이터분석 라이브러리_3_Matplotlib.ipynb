### Matplotlib
- 데이터 및 분석 결과를 그래프나 차트로 시각화하기 위한 기본적인 파이썬 라이브러리
- 공식 페이지 : https://matplotlib.org/stable/plot_types/index

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   #하위 모듈인 pyplot까지 접근
```

#### Plot chart
- 꺾은선 그래프

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y)       # plot : 꺾은선 그래프 출력 함수
plt.show()          # show : 차트를 화면에 표시하는 함수(쥬피터 노트북이 없어도 동작가능함.)
```

#### 차트 스타일 설정

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# ls : 라인 스타일 (Line style)
plt.plot(x,y, ls='--')
plt.show() 
```

#### Line width
- 선 굵기

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]


plt.plot(x,y, ls='--', lw = 0.8)
plt.show() 
```

#### Marker

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]


plt.plot(x,y, ls='--', lw = 0.8, marker='o')
plt.show() 
```

#### Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]


plt.plot(x,y, ls='--', lw = 0.8, marker='o', c='m')
plt.show() 
```

### 차트 옵션

#### 축 범위 지정

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', marker='o', lw = 5, ms=15, mec='g', mfc='r', mew =5)
plt.xlim(-2, 10)   # x축 범위     / lim : limit의 약자
plt.ylim(-5, 12)   # y축 범위
plt.show() 
```

#### 틱(tick) 설정
- 차트에서 축 상의 위치 표시 지점을 틱 이라고 함
- 틱에 써진 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
- 틱의 위치나 라벨은 Matplotlib에서 어느정도 자동으로 정해주지만 수동으로 설정도 가능

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', marker='o', lw = 5, ms=15, mec='g', mfc='r', mew =5)
plt.xticks([0,3,6])   # x 틱값 설정
plt.yticks([1,5,9])   # y 틱값 설정
plt.show() 
```

#### 그리드(grid)
- 차트의 틱 위치를 잘 보여주기 위해 차트 중간에 직선들을 그어주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', marker='o', lw = 5, ms=15, mec='g', mfc='r', mew =5)
plt.grid()     # 틱(tick)이 있는 지점만 grid가 생성됨
plt.show() 
```

#### X,Y축 및 그래프 제목

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls='--', marker='o', lw = 5, ms=15, mec='g', mfc='r', mew =5)
plt.xlabel("X axis")    # x축 제목
plt.ylabel("Y axis")    # y축 제목
plt.title("X-Y axis")   # 그래프 전체 제목
plt.show() 
```

### 여러 개 플롯 한번에 그리기

```python
x = np.arange(4)
y1 = [1,2,3,4]
y2 = [2,4,6,8]
y3 = [3,6,9,12]

# 그래프 전체 크기 조정 (행, 열)
plt.figure(figsize=(10,8))

plt.plot(x, y1, label='y1')   # label : 하나의 꺾은선 라인에 대한 라벨값 지정
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

# legend : 범례 설정
 # loc : 범례의 위치를 표시 (upper, center, lower) + (left, center, right)
 # prop : 범례 크기 조절
plt.legend(loc='upper right', prop = {'size':15})
plt.show()
```

### 바 차트(bar chart)
- 막대그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]
```

```python
plt.bar(x,y)
plt.show
```

#### 막대 색상 변경

```python
plt.bar(x,y, color = 'orange')
plt.show
```

```python
plt.bar(x,y, color = ['orange', 'cyan', 'pink'])        # 색상의 개수와 막대의 개수가 맞지 않으면 색상이 반복 적용됨.
plt.show
```

#### edgecolor
- 막대 테두리 색상 설정

```python
plt.bar(x,y, color = ['orange', 'cyan', 'pink'], edgecolor='blue') 
plt.show
```

#### linewidth
- 막대 테두리 굵기 설정

```python
plt.bar(x,y, color = ['orange', 'cyan', 'pink'], edgecolor='blue' , lw = 4) 
plt.show
```

#### alpha
- 투명도 설정

```python
plt.bar(x,y, color = ['orange', 'cyan', 'pink'], edgecolor='blue' , lw = 4, alpha=0.5)  # alpha는 0~1 사이값 
plt.show
```

#### width
- 막대 너비 설정

```python
plt.bar(x,y, color = ['orange', 'cyan', 'pink'], edgecolor='blue' , lw = 4, alpha=0.5, width=0.5) # width의 범위는 0~1 로 기존 너비의 비율을 지정
plt.show
```

### 수평 바 차트

```python
plt.barh(x,y)   # 수직 바 차트와는 달리 첫 번째 변수에 y축의 값들이 들어감
plt.show()
```

```python
# 수평 바 차트에서는 높이인 height(높낮이)로 막대의 너비를 결정함. 
plt.barh(x,y, color = ['orange', 'cyan', 'pink'], edgecolor='blue' , lw = 4, alpha=0.5, height=0.5)
plt.show
```

### subplot 활용
- 여러 개의 차트를 한 번에 표현하기

```python
x = ['group_A','group_B', 'group_C']
y = [1, 7, 15]
plt.figure(figsize=(9,3))

# subplot : 여러개의 하위 플롯을 생성하는 함수
plt.subplot(131)    # 131의 의미 : 전체 공간을 1행 3열로 분리시키고 그 중 1번째라는 뜻.
plt.bar(x,y)
plt.title('bar')

plt.subplot(132)
plt.scatter(x,y)
plt.title('scatter')

plt.subplot(133)
plt.plot(x,y)
plt.title('plot')

plt.suptitle('Subplot example', fontsize=15)
plt.tight_layout()   # 차트상 겹쳐지는 부분들을 적절히 조절해주는 기능
plt.show
```

#### 사고유형이 '차대차'인 사건 중에서 죽거나 다친 사람이 많은 '시도'를 알아보고 시각화 해보자

#### 중요한 키워드들이 들어있는 위치 확인
- 사고유형_대분류에서 ''차대차'인 행만 DF로 출력
- '차대차'인 행 중에서 '발생지시도'와 '사상자수' 컬럼만 확인

```python
ac = data[data['사고유형_대분류']=='차대차']
```

```python
ac.shape
# 사고유형이 차대차 인 사건은 총 1642건
```

```python
ac[['발생지시도','사상자수']]
```

```python
ac['발생지시도'].value_counts()
# 1642개의 사건 중에서 각 시도의 빈도수 출력
```

```python
ac_sum = ac[['발생지시도', '사상자수']]. groupby('발생지시도').sum()
ac_sum

# 차대차 사건 중에서 발생지시도 컬럼을 인덱스 컬럼으로 설정하여 내부값들(시도) 별로 사상자수의 합을 출력하는 코드
# 발생지시도와 사상자수만 골라내기,발생지시도 기준으로 같은 지역끼리 묶기, 묶인 지역별로 사상자수 더하기.
```

```python
# 중요해 ***

# x, y축 값 설정
x = np.arange(len(ac_sum.index))        # arange(17)써도 상관없음.
# y축은 위에서 구한 발생지시도별 사상자수 데이터를 내림차순으로 정렬하여 넣어줌.
y = ac_sum['사상자수'].sort_values(ascending=False)              # 솔트 언더바 밸류????
```

```python
# 그래프 그리기
plt.figure(figsize=(15,7))
plt.bar(x,y)
plt.xticks(x,y.index)
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.title('차vs차 교통사고의 사상자 수')
plt.show
```
