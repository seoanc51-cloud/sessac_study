```python
%%writefile module/myApp.py

import streamlit as st
import pandas as pd

# íƒ€ì´í‹€ ì¶œë ¥
st.title('*íƒ€ì´í‹€* -  ëŒ€') # * *ë§ˆí¬ë‹¤ìš´ íƒœê·¸ ê¸°í˜¸ë¡œ itablic(ê¸°ìš¸ì–´ì§)ì„ ì ìš©í•¨
st.header('íƒ€ì´í‹€ - ì¤‘')
st.subheader('íƒ€ì´í‹€ - ì†Œ')

# ì¼ë°˜ í…ìŠ¤íŠ¸ ì¶œë ¥
st.text('ê°„ë‹¨í•œ *í…ìŠ¤íŠ¸* ì¶œë ¥')   # * * ë§ˆí¬ë‹¤ìš´ íƒœê·¸ë¥¼ ì¸ì‹í•˜ì§€ ëª»í•¨

# ë§ˆí¬ë‹¤ìš´ ì¶œë ¥
st.markdown('*ë§ˆí¬ë‹¤ìš´* ì¶œë ¥')
st.markdown('**ë§ˆí¬ë‹¤ìš´** ì¶œë ¥')          # bold(ì§„í•˜ê²Œ) ì‚¬ìš©
st.markdown('***ë§ˆí¬ë‹¤ìš´*** ì¶œë ¥')        # itablic + bold ì ìš©
st.markdown('~~ë§ˆí¬ë‹¤ìš´~~ ì¶œë ¥')          # strikethrough(í…ìŠ¤íŠ¸ ì¤‘ê°„ ì¤„) ì ìš©
st.markdown('[ë‰´ìŠ¤ ì›ë¬¸ í´ë¦­](https://n.news.naver.com/mnews/article/020/0003647688)')     # í•˜ì´í¼ë§í¬ ì ìš©
st.markdown(':heart_eyes: :ocean: :stars:')     # ì´ëª¨ì§€ ì ìš©

# í…ìŠ¤íŠ¸ ë° ë³€ìˆ˜/ê°ì²´ ì¶œë ¥(íŒŒì´ì¬ì˜ printë¬¸ê³¼ ìœ ì‚¬í•¨)
st.write('í…ìŠ¤íŠ¸ë‚˜ íŒŒì´ì¬ ë³€ìˆ˜/ê°ì²´ë¥¼ ì¶œë ¥')
x, y = 1, 2                 # ê²‰ìœ¼ë¡œ ë“¤ì–´ë‚˜ëŠ” ì½”ë“œê°€ ì•„ë‹˜
st.write(x,y)
st.write('x=' , x, 'ê·¸ë¦¬ê³ ', 'y=', y)     # í¬ë§·íŒ…ì´ ì—†ëŠ” ê²½ìš° ìˆ«ìëŠ” ì´ˆë¡ìƒ‰ìœ¼ë¡œ ì¶œë ¥
st.write(f'x={x} ê·¸ë¦¬ê³  y={y}')           # í¬ë§·íŒ…ì´ ìˆëŠ” ê²½ìš° ì „ì²´ê°€ ë‹¤ ë¬¸ì(ê²€ì€ìƒ‰)ë¡œ ì¶œë ¥

df = pd.DataFrame(data={'ì»¬ëŸ¼1':[1,2,3],
                        'ì»¬ëŸ¼2':['a','b','c'],
                        'ì»¬ëŸ¼3':[True, False, True]
                        # TrueëŠ” ì›¹ ìƒì—ì„œ ì²´í¬í‘œì‹œë¡œ ëœ¨ê³ , FalseëŠ” ë¹ˆì¹¸ìœ¼ë¡œ ëœ¸
                       })
st.write(df)                    # ê¼­ ìŠ¤íŠ¸ë¦¼ì„ í•´ì¤˜ì•¼ í‘œì‹œê°€ ë¨.

# ìˆ˜ì‹ ì¶œë ¥
# late í•¨ìˆ˜ì—ì„œëŠ” '\'ëŠ” íŠ¹ì •í•œ ìˆ˜ì‹ì„ í‘œí˜„í•˜ëŠ” ê¸°í˜¸ì¸ë° íŒŒì´ì¬ ë¬¸ìì—´ ì´ìŠ¤ì¼€ì´í”„ ì½”ë“œë¡œ ì¸ì‹ë˜ì§€ ì•Šë„ë¡ ë¬¸ìì—´ ì•ì— r(raw stirng)ì„ ë¶™ì—¬ ë°©ì§€
st.latex('E = mc^2')
st.latex(r'Area = \pi r^2 \ ì…ë‹ˆë‹¤.')      # íŒŒì´ : \pi                     - ê·¼ë° ì—¬ê¸° ë’¤ì—ëŠ” ì™œ ë˜ \ ?
st.latex(r'\face{a}{b} = c')             # ë¶„ìˆ˜ : \frax{ë¶„ì}{ë¶„ëª¨}
st.latex(r'\sqrt{3}')                    # ì œê³±ê·¼ : \sqrt{í‘œí˜„ì‹}

# ì½”ë“œ ì¶œë ¥(ì½”ë“œ í˜•ì‹ìœ¼ë¡œ ì¶œë ¥í•˜ì§€ë§Œ ì‹¤ì œ ì½”ë“œ ë‚´ë¶€ ì—°ì‚°ì´ ì§„í–‰ë˜ì–´ ê²°ê³¼ê°€ ë‚˜ì˜¤ì§€ëŠ” ì•ŠìŒ)
myCode = '''
total = 0
for i in range(5):
        total += i
print(total)
'''
st.code(myCode)

# ìº¡ì…˜ ì¶œë ¥
st.caption('ì§§ì€ ì„¤ëª…ë¬¸(**ìº¡ì…˜**) ì¶œë ¥!')
```

```python
%%writefile module/myApp5.py

import streamlit as st
import pandas as pd

# ë²„íŠ¼ ê¸°ëŠ¥
 # ë˜‘ê°™ì€ labelë¡œ ì—¬ëŸ¬ê°œì˜ ë²„íŠ¼ì„ ë§Œë“¤ ìˆ˜ ì—†ìœ¼ë©° ê°™ì€ ë²„íŠ¼ì„ ë§Œë“œë ¤ë©´ key ë§¤ê°œë³€ìˆ˜ë¡œ êµ¬ë¶„ì‹œì¼œì¤˜ì•¼í•¨
st.button(label='í´ë¦­!', key=1)
st.button(label='í´ë¦­!', key=2, help='ì´ê²ƒì€ :blue[**íˆ´íŒ**]')

# ë²„íŠ¼ì„ ëˆŒë €ì„ ë•Œ ë™ì‘í•  ê¸°ëŠ¥ ì„¤ì •
def myFunc(*args) :
    total = 0
    for i in args :
        total += i
    st.write(f'í•©ê³„ : {total}')

# on_click : ë²„íŠ¼ì„ ëˆŒë €ì„ ë•Œ ë™ì‘í•  í•¨ìˆ˜ë¥¼ ì„¤ì •í•˜ëŠ” ë§¤ê°œë³€ìˆ˜(ë’¤ì˜ argsì˜ ë¦¬ìŠ¤íŠ¸ì— ìˆëŠ” ê°’ë“¤ì´ myFuncì˜ ë§¤ê°œë³€ìˆ˜ë¡œ ë“¤ì–´ê°)
st.button('í´ë¦­!' , key=3, on_click=myFunc, args=[1,2,3,4,5])

button = st.button('í´ë¦­!', key=4)
st.write(button)      # í´ë¦­ ì „ì—ëŠ” Falseì´ê³  í´ë¦­í•˜ë©´ Trueë¡œ ë³€ê²½
if button :
    st.write(':smile:')
else :
    st.write(':sunglasses:')

''
'___'
''

# ì²´í¬ë°•ìŠ¤
temp = st.checkbox('ìœ„ ë‚´ìš©ì— ë™ì˜í•©ë‹ˆë‹¤.')
st.write(temp)
''
'___'
''

# ë¼ë””ì˜¤ ë°•ìŠ¤
radio = st.radio(label='ë‹¤ìŒ ì¤‘ í•œê°€ì§€ ì„ íƒ', options=['ë¶€ë¨¹', 'ì°ë¨¹'], key='a')   # ìœ„ì ¯ì´ ë‹¬ë¼ë„ keyê°’ì´ ì¤‘ë³µë˜ë©´ ì—ëŸ¬ê°€ ë°œìƒ
st.write(f'**radio ì„ íƒ =** {radio}')
''
'___'
''
# ì„ íƒ ë°•ìŠ¤ (ì—¬ëŸ¬ê°œì˜ ê°’ ì¤‘ í•˜ë‚˜ ì„ íƒ)
select = st.selectbox(label='ë‹¤ìŒ ì¤‘ í•œê°€ì§€ ì„ íƒ', options=['ë¶€ë¨¹', 'ì°ë¨¹'])
st.write(f'**selectbox ì„ íƒ =** {select}')
''
'___'
''
# ë‹¤ì¤‘ ì„ íƒ ë°•ìŠ¤ (ì„ íƒ ë°•ìŠ¤ë‘ ìœ ì‚¬í•˜ë‚˜ ì—¬ëŸ¬ ê°œì˜ ê°’ì„ ë™ì‹œì— ì„ íƒ ê°€ëŠ¥)
mul_sel = st.multiselect('ì˜¤ëŠ˜ ì €ë… ë©”ë‰´ëŠ”!?', ['ë§ˆëŠ˜ì¹˜í‚¨', 'ë§ˆë¼íƒ•', 'ëˆê°€ìŠ¤&ëƒ‰ëª¨ë°€', 'ì—½ë–¡'])
st.write(f'**mul_sel ì„ íƒ =** {mul_sel}')   # ë‹¤ì¤‘ ì„ íƒ ë°•ìŠ¤ë¡œ ì„ íƒí•œ ê°’ë“¤ì€ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥ë¨
''
'___'
''
# ì»¬ëŸ¬ ì„ íƒ (ë³€ìˆ˜ ì¶œë ¥ì‹œ HEXì»¬ëŸ¬ì½”ë“œ ê°’ìœ¼ë¡œ ì¶œë ¥ë¨)
c = st.color_picker('ì»¬ëŸ¬ ì„ íƒ')
st.write(f'color picker ì„ íƒ = {c}')
```

```python
%%writefile module/myApp6.py

import streamlit as st
import pandas as pd
import datetime

# ì„ íƒ ìŠ¬ë¼ì´ë”(ì •í•´ì§„ ê°’ ì¤‘ì—ì„œ ì„ íƒ)
sel_slider = st.select_slider('í•´ë‹¹ í˜ì´ì§€ë¥¼ ì¶”ì²œí•˜ì‹œë‚˜ìš”?', ['ë§¤ìš° ì¶”ì²œ', 'ì¶”ì²œ', 'ë³´í†µ', 'ë¹„ì¶”ì²œ', 'ë§¤ìš° ë¹„ì¶”ì²œ'])
st.write(f'ì¶”ì²œë„ = {sel_slider}')
''
'___'
''
# ìŠ¬ë¼ì´ë”(ì •ìˆ˜ ì…ë ¥)
 # min_value : ìµœì†Œê°’, max_value : ìµœëŒ€ê°’, value : ì´ˆê¸°ê°’, step : ìµœì†Œ ë‹¨ìœ„(ë””í´íŠ¸ 1)
slider = st.slider('ì¶”ì²œ ì ìˆ˜ëŠ”?', min_value=0, max_value=10, value=5, step=2)
st.write(f'ì¶”ì²œì ìˆ˜ = {slider}')
''
'___'
''
# ìˆ«ì ì…ë ¥
num_input = st.number_input('ì´ë²ˆ ì‹œí—˜ ì ìˆ˜ëŠ”?', min_value=0, max_value=100, value=0, step=5)
st.write(f'ì‹œí—˜ ì ìˆ˜ = {num_input}')
''
'___'
''
# ê°ì¢… ì…ë ¥ ìœ„ì ¯
 # í…ìŠ¤íŠ¸ ì…ë ¥
txt_input = st.text_input('ë‹¹ì‹ ì˜ ì´ë¦„ì€? :sunglasses:')
 # ë‚ ì§œ ì…ë ¥
 # datetime.now : í˜„ì¬ ë…„ì›”ì¼ì‹œ
 # max_valueì„¤ì •ì´ ì—†ë‹¤ë©´ í˜„ì¬ ì‹œì ìœ¼ë¡œë¶€í„° -10ë…„, ~ +10ë…„ê¹Œì§€ì˜ ë””í´íŠ¸ë¡œ ì…ë ¥ê°€ëŠ¥í•¨
d_input = st.date_input('ë‹¹ì‹ ì˜ ìƒì¼ì€?', max_value=datetime.datetime.now())
 # ì‹œê°„ ì…ë ¥
t_input = st.time_input('ë‹¹ì‹ ì´ íƒœì–´ë‚œ ì‹œê°„ì€?' , step=datetime.timedelta(hours=1))  # seconds, minutes, hoursë¡œ êµ¬ê°„ ì„¤ì • ê°€ëŠ¥
st.write(f'ì´ë¦„={txt_input}, ìƒì¼={d_input}, ì¶œìƒ ì‹œê°„={t_input}')
```
